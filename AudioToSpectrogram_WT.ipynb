### Perform the audio (WAV) -> FT spectrogram (PNG) for all files

```python
filename_list = []
coikw = {'alpha': 0.5, 'hatch': '/'} 
```

```python
for emotionpath in os.listdir(path_to_wavfiles):
    for file in glob.glob(path_to_wavfiles + '\\' + emotionpath + '/**/*.wav', recursive=True):
        filename_list.append(file)
        if not exists(file[:-3]+"png"):
            
            # Get samples
            print("Processing " + file)
            sample_rate, samples = wavfile.read(file)
            frequencies, times, spectrogram = signal.spectrogram(samples, sample_rate)
            
            # Plot
            ax2 = scg.cws(samples, scales=scales, figsize=(6.9,2.3), coikw=coikw); 
            fig=plt.gcf()
            fig.tight_layout()
            
            # Save file
            print("Creating " + file[:-3]+"png")
            fig.savefig(file[:-3]+"png", format = 'png', dpi=300)
            
            # Free fig mem
            plt.clf()
            plt.close(fig)
        else:
            print("Skipping " + file)
```
